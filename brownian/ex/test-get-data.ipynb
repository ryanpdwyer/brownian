```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline

import jittermodel as jm
import brownian
```

```python
f, PSD_mean, PSD_ci = brownian.get_data('../brownianmotion10x164357.h5')
```

```python
from brownian import u
cantilever_guess = brownian.UnitCantilever(f_c=60*u.kHz, k_c = 3.5*u.N/u.m, Q=25000*u.dimensionless)
```

```python
m = brownian.make_mask(f, 59900, 60100)
plt.semilogy(f[m], PSD_mean[m])
bmf = brownian.BrownianMotionFitter(f[m], PSD_mean[m], PSD_ci[m], 298, cantilever_guess)
```

```python
bmf.calc_fit(59900, 60100)
```

```python
print("{self.f_min:.0f} {self.f_max:.0f}".format(self=bmf))
bmf.plot_fit()
bmf.print_output()
```
